```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
square = np.genfromtxt('square.csv', delimiter=',')[1:]
wide = np.genfromtxt('wide.csv', delimiter=',')[1:]
tall = np.genfromtxt('tall.csv', delimiter=',')[1:]
```

```python
# Square Matrix
plt.plot(square[:,0], square[:,1:3], label=['row_wise', 'col_wise'])
plt.legend()
```

```python
# Wide Matrix
plt.plot(wide[:,0], wide[:,1:3], label=['row_wise', 'col_wise'])
plt.legend()
```

```python
# Tall Matrix
plt.plot(tall[:,0], tall[:,1:3], label=['row_wise', 'col_wise'])
plt.legend()
```
